# Exploratory Data Analysis in Pandas

### Project Goal: Perform exploratory data analysis to gather valuable insights into a dataset containing world population data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\majorlydata\Downloads\world_population.csv")
df
```

```python
# Re-format float numbers
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Find sum of null values within dataset
df.isnull().sum()
```

```python
# Show count of unique values in each column
df.nunique()
```

```python
# Show bottom 10 countries by 2022 population
df.sort_values(by = "2022 Population", ascending = True)[["Country", "2022 Population"]].head(10)
```

```python
# Show top 10 countries by area
area_col = df.filter(like = 'Area').columns[0]
df.sort_values(by = area_col, ascending = False)[["Country", area_col]].head(10)
```

```python
# Make a heatmap to show the correlation between columns
sns.heatmap(df.corr(numeric_only = True), annot = True)

plt.figure(figsize = (10,7))

plt.show()
```

### Plot average continent population data

```python
cont_avg_pop = df.groupby('Continent')[df.columns[5:13][::-1]].mean(numeric_only = True).reset_index() \
.sort_values(by = '2022 Population', ascending = False)
cont_avg_pop
```

```python
# Configure continent average population data frame
cont_avg_pop_melted = pd.melt(cont_avg_pop, id_vars = 'Continent', var_name = 'Year', value_name = 'Avg Population')
```

```python
# Plot the data
sns.lineplot(data = cont_avg_pop_melted, x = 'Year', y = 'Avg Population', hue = 'Continent', marker = 'o')

plt.xticks(rotation = 90)
plt.grid(axis = 'y', linestyle = '--')
plt.tight_layout()
plt.figure(figsize = (10,7))
plt.show()
```
